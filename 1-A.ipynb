```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
```

```python
#setting the shape parameters
#a = alpha (successes) b = beta (failures)
a,b = 2,2

prior = np.linspace(beta.ppf(0.01, a, b),
                beta.ppf(0.99, a, b), 100)
```

```python
#plotting

plt.figure(figsize=(3,3))
plt.xlim(0, 1)
plt.plot(prior, beta.pdf(prior, a, b))
plt.title('Prior Beta Distribution')
plt.xlabel('Values of Random Variable X (0, 1)')
plt.ylabel('Probability')
plt.show()
```

```python
#posterior after survey 1

a,b = 42,12
posterior = np.linspace(beta.ppf(0.01, a, b),
                beta.ppf(0.99, a, b), 100)
```

```python
plt.figure(figsize=(3,3))
plt.xlim(0, 1)
plt.plot(posterior, beta.pdf(posterior, a, b))
plt.title('Posterior Beta Distribution after Survey 1 or Prior Before Survey 2')
plt.xlabel('Values of Random Variable X (0, 1)')
plt.ylabel('Probability')
plt.axvline(0.7884,  linestyle = 'dashed', color='black')
plt.show()
```

```python
#posterior after survey 2

a,b = 55,29
posterior = np.linspace(beta.ppf(0.01, a, b),
                beta.ppf(0.99, a, b), 100)
```

```python
plt.figure(figsize=(3,3))
plt.xlim(0, 1)
plt.plot(posterior, beta.pdf(posterior, a, b))
plt.title('Posterior Beta Distribution after Survey 2')
plt.xlabel('Values of Random Variable X (0, 1)')
plt.ylabel('Probability')
plt.axvline(0.658,  linestyle = 'dashed', color='black')
plt.show()
```

i) The prior beta distribution for survey 2 will be the same as the posterior distribution of the data post survey 1.
This is because the prior distribution is defined as the probability distribution without taking into account the evidence (results of the 2nd survey).

```python
#posterior after survey 3

a,b = 125,59
posterior = np.linspace(beta.ppf(0.01, a, b),
                beta.ppf(0.99, a, b), 100)
```

```python
plt.figure(figsize=(3,3))
plt.xlim(0, 1)
plt.plot(posterior, beta.pdf(posterior, a, b))
plt.title('Posterior Beta Distribution after Survey 3')
plt.xlabel('Values of Random Variable X (0, 1)')
plt.ylabel('Probability')
plt.axvline(0.6813,  linestyle = 'dashed', color='black')
plt.show()
```

```python
#Likelihood function

aLH = 40+13+70+1
bLH = 10+17+30+1
likelihood = np.linspace(beta.ppf(0.01,aLH,bLH), beta.ppf(0.99, aLH, bLH), 100)
plt.figure(figsize=(3,3))
plt.xlim(0, 1)
plt.plot(likelihood, beta.pdf(likelihood, aLH, bLH))
plt.title('Likelihood of s')
plt.xlabel('Values of Random Variable X (0, 1)')
plt.ylabel('Probability')
plt.show()
```

ii) The posterior distribution of x is defined as the updation of the prior distribution of x with the new evidence to give a revised probability distribution of an event.
The posterior distribution of s gathered by incrementally updating the prior distribution (a beta distribution of parameters 2,2) with the results of the survey at each point (3 survey updates).
The results of the survey are added to the initial $\alpha$ and $\beta$ values to give the final posterior probability distribution as a beta distribution with parameters $\alpha$ = 125, $\beta$ = 59.
